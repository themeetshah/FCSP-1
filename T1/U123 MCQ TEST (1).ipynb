### Q6. What will be the output of the following program?

```python
x = 0
count = 0
for x in range(10):
    while x < 15:
        if x < 0:
            pass
        elif x % 2 == 0:
            x += 1
            continue
        elif x % 3 == 0:
            x += 1
            continue
        elif count == 5:
            break
        count += 1
print(x, count)
```

- [ ] A) Infinite loop
- [ ] B) `x = 11, count = 5`
- [ ] C) `x = 15, count = 10`
- [ ] D) `None`
