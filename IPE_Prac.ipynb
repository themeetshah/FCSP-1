```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
workout_durations_data = [40, 50, 45, 55, 60, 30, 40, 50, 45, 55, 60, 30, 40, 50, 45, 55, 60, 30, 40, 50, 45, 55, 60, 30, 40, 50, 45, 55]
calories_burned_data= [200, 250, 220, 270, 300, 150, 200, 250, 220, 270, 300, 150, 200, 250, 220, 270, 300, 150, 200, 250, 220, 270, 300, 150, 200, 250, 220, 270]

duration_weekly = []
calories_weekly = []
rate_of_change = []

for i in range(0, len(workout_durations_data), 7):
    duration_weekly.append(np.mean(workout_durations_data[i:i+7]))
    calories_weekly.append(np.mean(calories_burned_data[i:i+7]))

for i in range(0,len(workout_durations_data)-1):
    rate_of_change.append(np.abs(workout_durations_data[i+1]-workout_durations_data[i]))
# print(duration_weekly)
# print(duration_weekly)

days = np.arange(1, len(workout_durations_data) + 1)
weeks = np.arange(1, len(duration_weekly) + 1)

plt.subplot(3,2,1)
plt.title("Daily Workout Durations")
plt.plot(days, workout_durations_data, color='blue', marker='o', ms=8)
plt.xlabel("Days")
plt.ylabel("Duration (in Minutes)")

plt.subplot(3,2,2)
plt.title("Weekly Averages")
plt.plot(weeks, duration_weekly, color='green', marker='o', ms=8, label='Workout duration')
plt.plot(weeks, calories_weekly, color='orange', marker='s', ms=8, label='Calories burned')
# plt.legend(loc="upper right", bbox_to_anchor=(3, 1))
plt.legend()
plt.xlabel("Weeks")
plt.ylabel("Average values")

plt.subplot(3,2,3)
plt.title("Daily Workout Durations")
plt.bar(days, workout_durations_data, color='blue')
plt.xlabel("Days")
plt.ylabel("Duration")

plt.subplot(3,2,4)
plt.title("Workout Durations Distribution {Histogram}")
plt.hist(workout_durations_data, color='skyblue', bins=10)
plt.xlabel("Duration")
plt.ylabel("Frequency")

plt.subplot(3,2,5)
plt.title("Calories Burned Distribution (Pie Chart)")
plt.pie(calories_weekly, labels=weeks, startangle=90, autopct='%1.1f%%')

plt.subplot(3,2,6)
plt.title("Rate of Change in Workout Durations")
plt.plot(rate_of_change, color='red', marker='^', ms=8)
# print(rate_of_change)

plt.suptitle("Workout Analysis")

plt.tight_layout(w_pad=5.0)
plt.show()
```

```python
Scores = [[31, 12, 19, 53], 
          [67, 48, 95, 83], 
          [59, 67, 13, 59], 
          [62, 29, 99, 88], 
          [87, 91, 69, 76]]

scores = np.array(Scores)

print("the maximum score in T_20-3:",np.max(scores[:,2]))

print("the minimum score of YUVRAJ:",np.min(scores[2]))

batsman_total = []
for i in range(len(scores)):
    add=0
    for j in range(len(scores[i])):
        add+=scores[i,j]
    batsman_total.append(add)
batsman_total = np.array(batsman_total).reshape(len(scores),1)
print("after adding:")
print(np.concatenate((scores,batsman_total), axis=1))

print()

print("sorted:")
print(scores[np.argsort(np.concatenate((scores,batsman_total), axis=1)[:,4])[::-1]])

print()

match_total = []
for i in range(len(scores)-1):
    match_total.append(np.sum(scores[:,i]))
match_total = np.array(match_total).reshape(1,len(scores)-1)
print("match:")
print(np.concatenate((scores,match_total), axis=0))

print()

print("greater than 30:")
print(np.where(scores>30,1,0))

print()

plt.hist(scores[0], bins=[0,20,40,60])
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```

```python
stack = []
class SQ:
    def __init__(self, L):
        self.L = L

    def shift(self):
        try:
            if len(self.L) == 0:
                raise Exception('List is empty')
            return self.L.pop(0)
        except Exception as e:
            print(e)

    def unshift(self, n):
        self.L.insert(0,n)

    def push(self, n):
        self.L.append(n)

    def pop(self):
        try:
            if len(self.L) == 0:
                raise Exception('List is empty')
            return self.L.pop()
        except Exception as e:
            return e

    def remove(self):
        try:
            if len(self.L) == 0:
                raise Exception('List is empty')
            return self.L.remove(max(self.L))
        except Exception as e:
            return e

    def display(self):
        return self.L

sq = SQ([1,4,6,8,9])
print(sq.display())
print(sq.shift())
print(sq.display())
print(sq.unshift(7))
print(sq.display())
print(sq.push(10))
print(sq.display())
print(sq.remove())
print(sq.display())
print(sq.pop())
print(sq.display())
print(sq.pop())
print(sq.display())
print(sq.pop())
print(sq.display())
print(sq.pop())
print(sq.display())
print(sq.pop())
print(sq.display())
print(sq.pop())
print(sq.display())
print(sq.pop())
print(sq.display())
```

```python
string="Tit omlahshagiagi o rnnespnpagg y,rm u"
key=5
length=len(string)#to get the length of the input
print(length)
part=length//key #half of the input
print(part)
extra=length%key #extra characters after dividing string in equal parts
print(extra)
part1=string[:(part+1)*extra]
part2=string[(part+1)*extra:]
print(part1)
print(part2)
msg=''
for i in range(part+1):
    if i<part:
        msg+=part1[i::part+1]+part2[i::part]
    else:
        msg+=part1[i::part+1]
print(msg)
```

```python
n=int(input("enter value of n (n*n matrix)"))
matrix=[]
for i in range(0,n):
    matrix.append([])
    for j in range(0,n):
        matrix[i].append(int(input(f"enter element of {i,j}: ")))
print(matrix)
for i in range(0,n):
    for j in range(0,n):
        if i>j:
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

for r in range(len(matrix)):
    matrix[r].reverse()
print(matrix)
```
