#### 11. What will be the output of the following Python code? 
```python
L1 = [1, 1.33, 'LJU', 0, 'N', True, 'Y', 1] 
val1 = 0 
val2 = "" 
for x in L1: 
    if type(x) == int or type(x) == float: 
        val1 += x 
    elif type(x) == str: 
        val2 += x 
    else: 
        break 
        continue 
print(val1, val2) 
```

- c) `2.33 LJUN`  
- d) `Error`

#### 12. What will be the output of the following Python code? 
```python
def sum_list(l): 
    sum = 0 
    for i in range(len(l)): 
        if l[i] == 13 or l[i - 1] == 13: 
            continue 
        else: 
            sum += l[i] 
    return sum    
l = [1, 2, 13, 2, 9, 13] 
print(sum_list(l)) 
```
- a) `11`  
- b) `12`  
- c) `15`  
- d) `Error`
