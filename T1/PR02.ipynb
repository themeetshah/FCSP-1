```python
#A
n = 5  # Height of the letter A
for i in range(n):
    for j in range(n):
        # Print spaces for the left side of the A
        if j < n - i - 1:
            print(' ', end="")
        # Print the left leg of A
        elif j == n - i - 1:
            print('*', end=" ")
        # Print the right leg of A
        elif j == n + i:
            print('*', end=" ")
        # Print the middle bar of A
        elif i == n // 2 and j > n - i - 1 and j < n + i:
            print('*', end=" ")
        else:
            print(' ', end="")
    print()  # Move to the next line after each row
```

```python
#wap to find the sum of max 2 nos out of 5 nos entered by the user

highest=None
second_highest=None
for i in range(1,6):
    a=int(input(f"Enter num{i}: "))
    if not(a==highest or a==second_highest):
        if(highest==None):
            highest=a
            second_highest=a
        if(a>highest):
            second_highest=highest
            highest=a
        elif(a>second_highest):
            second_highest=a
    else:
        print("Not duplicates")
        break
else:
    print(highest+second_highest)
```
