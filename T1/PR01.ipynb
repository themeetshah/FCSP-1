```python
# given an int return the number of steps to reduce it to zero. if the current num is even divide it by 2 else subtract 1 from it
num = int(input("enter num: "))
steps=0
while(num!=0):
    if(num%2==0):
        num//=2
    else:
        num-=1
    steps+=1
print(steps)
```

```python
#a collatz sequence is generated by applying following rules in terms:
#    if even, divide by 2
#    if odd, multiply by 3 and add 1
#the algo says that it always reaches 1 for +ve int
#create function for 2 +ve int a and b and returns a if int a took fewer steps to reach 1 than b

def func(num):
    steps=0
    while(num!=1):
        if(num%2==0):
            num//=2
        else:
            num*=3
            num+=1
        steps+=1
    return steps

def who(a,b):
    a_step=func(a)
    b_step=func(b)
    if a_step>b_step:
        return b, b_step
    elif a_step<b_step:
        return a, a_step
    else:
        return a, a_step
    
a = int(input("enter a: "))
b = int(input("enter b: "))
print(who(a,b))
```

```python
# num 99 has a property 9*9=81 and 9+9=18 so 18+81=99
def chk(num):
    temp=num
    product=1
    sum1=0
    while(temp>0):
        r=temp%10
        product*=r
        sum1+=r
        temp//=10
    if(sum1+product==num):
        print(num)
        
for i in range(1,10001):
    chk(i)
```
